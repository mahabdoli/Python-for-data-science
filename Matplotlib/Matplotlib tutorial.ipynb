```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Matplotlib has two interfaces:
- based on MATLAB and uses a state-based interface
- an object-oriented (OO) interface

# Pyplot Interface
matplotlib on the surface is made to imitate MATLAB's method of generating plots, which is called pyplot. All the pyplot commands make changes and modify the same figure. This interface allows us to quickly and easily generate plots and allows us to add elements and/or modify the plot.

## Basic Examples


```python
#generating the data
import numpy as np
x = np.linspace(0, 10, 11)
y = x ** 2
```

```python
print(x)
print(y)
```

```python
# Plot y versus x as lines and/or markers.
plt.plot(x,y, 'b--') #'b--' sets the color to be blue, and the line style to be a dashed line
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X Squared')
```

```python
#you can change the color, style and width of the plot line.
print(x)
print(y)
plt.plot(x,y, color='g', linestyle='-.', linewidth=4)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X Squared')
```

```python
#you can plot two plots in the same figure
plt.plot(x,y)
plt.plot(y,x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X Squared')
```

## Creating Multiple plots 

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(2,2,1)
plt.plot(x, y, 'r--') # More on color options later
plt.xlabel('x, first plot')
plt.ylabel('y, first plot')
plt.subplot(2,2,2)
plt.plot(y, x, 'g*-')
plt.subplot(2,2,3)
plt.plot(x, y, 'k--')
plt.xlabel('x, third plot')
plt.ylabel('y, third plot')
plt.subplot(2,2,4)
plt.plot(y, x, 'b-.')
plt.tight_layout()
```

# Matplotlib - Object-oriented Interface
In this method the final plot that we see as the output is a Figure object. First, we create a figure object and then call methods or attributes.

```python
#first, create Figure
fig = plt.figure()
print(type(fig))

## Add an axes at position [*left*, *bottom*, *width*,*height*] where all quantities are in fractions of figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 

# Plot on axes
axes.plot(x, y, 'g-.')
axes.set_xlabel('X') 
axes.set_ylabel('X Squared')
axes.set_title('X Squared Plot')
```

```python
#first, create Figure
fig = plt.figure(figsize=(8,4))
print(type(fig))

## Add an axes at position [*left*, *bottom*, *width*,*height*] where all quantities are in fractions of figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 

# Plot on axes
axes.plot(x, y, 'g-.')
axes.set_xlabel('X') 
axes.set_ylabel('X Squared')
axes.set_title('X Squared Plot')
```

### Legends
You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure.<br>
The legend function takes an optional keyword argument loc that can be used to specify where in the figure the legend is going to be placed.Check the following link for more details:
https://matplotlib.org/tutorials/intermediate/legend_guide.html

```python
#create figure
fig = plt.figure()

#add axes to the figure object
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

#plot on axes
axes.plot(x, y, color='y', linestyle='-.', linewidth=4, label='plt1')
axes.plot(y,x, label='plt2')
axes.set_xlabel('X') 
axes.set_ylabel('Y')
axes.set_title('Two plots!')
axes.legend(loc=6)
```

## Add two axes

```python
# Creates figure
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

## changing Colors and line styles

You can change the colors of you plot in two ways:
- basic MATLAB like syntax
- using parameters 

### basic MATLAB like syntax
Here we use a string which consists of a part for color, marker and line. Here `'b'` means blue, `'g'` means green and so on. You can check the following link for more details: <br>
https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot


```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-.') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

## Changing colors and style using parameters
We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the color and alpha keyword arguments. Alpha indicates opacity.

```python
fig, ax = plt.subplots()

ax.plot(x, y, color='#ff0080', linestyle='--', linewidth=4, alpha=1, marker='o', markersize=10)
ax.plot(x, y-20, color='#8000ff',linestyle='-', linewidth=3, alpha=0.5, marker='*', markersize=15)
ax.plot(x, y-40, color='#00ff00',linestyle='--', linewidth=5, alpha=0.2, marker='+', markersize=20)

```

## Plot range
We can configure the ranges of the axes using the set_ylim and set_xlim methods in the axis object.

```python
fig, ax = plt.subplots()

ax.plot(x,y)
ax.set_xlim(0,50)
ax.set_ylim(0,77)
```

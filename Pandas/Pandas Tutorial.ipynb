## Multiple conditions using And and Or
In pandas you can't use python's built-in keywords for and and or instead use **&** and **|**.

```python
filt = (df['Director'] == 'David Yates') & (df['Runtime (Minutes)']>135)
df.loc[filt]
```

```python
filt = (df['Director'] == 'David Yates') & ((df['Rating']>7.5) | (df['Metascore']>70))
df.loc[filt]
```

## Add column using .apply() method
.apply() Applies a function along an axis of the DataFrame.
using .apply() method we are going to add a new column to our dataframe with True when a movie is longer than two hours and False when a movie is shorter.
first we have to define a function. 

```python
def longer(x):
    return x>120
```

```python
df['runtime_(minutes)'].apply(longer)
```

### lambda function
A lambda function is a small anonymous function (a function without a name). lambda functions have the below syntax:

    lambda arguments: expression

Lambda functions can be used when a function object is required.

```python
df['runtime_(minutes)'].apply(lambda x: True if x>=120 else False)
```

```python
df['longer_than_2'] = df['runtime_(minutes)'].apply(lambda x: True if x>=120 else False)
```

```python
df.head()
```

## Handling Custom missing values in .csv files
In addition to .replace() method introduced in pandas snippets file, before loading a .csv file you can specify a list containg custome values you want to be considered as nan values. pass the list to na_values argument in 

    nan_vals = ['Missing', 'no data']
    pd.read_csv('datasets_1474_2639_IMDB-Movie-Data.csv', na_values=nan_vals).
    
na_values: scalar, str, list-like, or dict, optional Additional strings to recognize as NA/NaN. If dict passed, specific per-column NA values.
